```python
import numpy as np 
from numpy import sin, cos
from scipy.integrate import odeint
from matplotlib import pyplot as plt 




# define the equations
def equations(y0, time):
	theta, x = y0
	f = [x, -(g/l) * sin(theta)]
	return f

def plot_results(time, theta2):
	#plt.plot(time, theta1[:,0])
	plt.plot(time, theta2)

	s = '(Initial Angle = ' + str(initial_angle) + ' degrees)'
	plt.title('Pendulum Motion: ' + s)
	plt.xlabel('time (s)')
	plt.ylabel('angle (rad)')
	plt.grid(True)
	plt.legend(['linear'], loc='lower right')
	plt.show()

# parameters
g = 9.81
l = 1.0

time = np.arange(0, 10.0, 0.025)
print(len(time))

# initial conditions
initial_angle = 30.0
theta0 = np.radians(initial_angle)
x0 = np.radians(0.0)

print('Anfangswinkel:', initial_angle, 'deg')
print('Anfangswinkel:', theta0, 'rad')
print('vel:', x0, 'rad/s?')

# find the solution to the nonlinear problem
theta1 = odeint(equations, [theta0, x0],  time)
print([theta0, x0])
print('theta1 len:', len(theta1))

# find the solution to the linear problem
w = np.sqrt(g/l)
theta2 = [theta0 * cos(w*t) for t in time]
print(type(theta2))

# plot the results
#plot_results(time, theta1, theta2)
plot_results(time, theta2)
```
